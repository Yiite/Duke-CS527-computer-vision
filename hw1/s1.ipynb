### Problem 1.1

$$

### Problem 1.4

Give expressions for all the unit-norm eigenvectors $\mathbf{u}_1$ of $H_f$ corresponding to the eigenvalue $\lambda_1 (x, y)$ you found in the previous problem and for all the unit-norm eigenvectors $\mathbf{u}_2$ corresponding to the eigenvalue $\lambda_2(x, y)$.

### Answer

The eigenvector system of equations corresponding to eigenvalue $\lambda_i$ (for $i = 1, 2$) is

$$
H \mathbf{u} = \lambda_i \mathbf{u}\;.
$$

When $i=1$, this system becomes (after dividing both sides by $6$)

\begin{eqnarray*}
(x-1) u &=& (x-1) u\\
(y-1) v &=& (x-1) v\;.
\end{eqnarray*}

The first equation is always true, and therefore if $\mathbf{u} = (u, v)^T$ then $u$ is unconstrained. The second equation admits any solution $v$ if $x = y$ and only solution $v = 0$ otherwise. In conclusion,

$$
\mathbf{u}_1 = \left\{\begin{array}{ll}
\left[\begin{array}{c}\cos\alpha \\ \sin\alpha \end{array}\right] &
\mbox{for any $\alpha\in [0,2\pi)$ if $x = y$} \\
\pm \left[\begin{array}{c} 1 \\ 0 \end{array}\right] & \mbox{otherwise}\;.
\end{array}\right.
$$

The reasoning for $\mathbf{u}_2$ is the same after simultaneously switching $x$ with $y$ and $u$ with $v$, so that

$$
\mathbf{u}_2 = \left\{\begin{array}{ll}
\left[\begin{array}{c}\cos\alpha \\ \sin\alpha \end{array}\right] &
\mbox{for any $\alpha\in [0,2\pi)$ if $x = y$} \\
\pm \left[\begin{array}{c} 0 \\ 1 \end{array}\right] & \mbox{otherwise}\;.
\end{array}\right.
$$

_**Notation:**_ There are many correct ways to say "any unit-norm vector" (including "any unit-norm vector"), and all these earn full credit.

### Problem 1.5

Give _all_ the stationary points of $f$. For each of them, state whether it is a minimum, a maximum, or a saddle point, and give unit vectors along the directions of maximum and minimum curvature (or state that curvatures in all directions are equal, if that is the case).

### Answer

Since $f$ is differentiable everywhere, the stationary points are the solutions of the system

$$
\nabla f(\mathbf{x}) = \mathbf{0} \;,
$$
that is,
$$
\begin{eqnarray*}
x(x - 2) &=& 0 \\
y(y - 2) &=& 0 \;.
\end{eqnarray*}
$$

There are four solutions to this system. We reason explicitly about the first, and then give answers for all of them.

One stationary point is $\mathbf{x}_1 = (0, 0)^T$.
    The Hessian there is
    $$
        H_f(\mathbf{x}_1) = -6 \left[\begin{array}{rr}1 & 0\\
        0 & 1\end{array}\right]\;.
    $$
    
Its two eigenvalues $\lambda_1 = \lambda_2 = -6$ are negative, so that the Hessian is negative definite, and $\mathbf{x}_1$ is a _maximum_. Since $x = y$ at this point, all nonzero vectors are eigenvectors, and all curvatures are the same ($-6$) at $\mathbf{x}_1$.

Similar reasoning leads to the following conclusions:

$$
\begin{array}{cccc}
\mathbf{x}_k & \text{Type} & \text{Dir of max curvature} &
\text{Dir of min curvature} \\\hline
(0, 0)^T & \text{maximum} & \text{any} & \text{any} \\
(2, 0)^T & \text{saddle} & (\pm 1, 0)^T & (0, \pm 1)^T \\
(0, 2)^T & \text{saddle} & (0, \pm 1)^T & (\pm 1, 0)^T \\
(2, 2)^T & \text{minimum} & \text{any} & \text{any}
\end{array}
$$

### Problem 3.3

What is the mean $m_{Y|X<2} = \mathbb{E}(Y\,|\, X < 2)$ of $Y$ given that $X < 2$?

### Answer

By definition of conditional expectation,
$$
m_{Y|X<2} \;=\; \mathbb{E}(Y\,|\, X < 2) \;=\;
\int_{\infty}^{\infty} y\, f_{Y|X<2}(y)\, dy
$$
and
\begin{eqnarray*}
f_{Y|X<2}(y) &=& \frac{f(y \;\text{and either}\; X = 0 \;\text{or}\; X = 1)}{p_0 + p_1} \;=\;
\frac{f((y \;\text{and}\; X = 0) \;\text{or}\; (y \;\text{and}\; X = 1))}{p_0 + p_1}\;.
\end{eqnarray*}
The two events at the numerator are incompatible, so their densities add up:
\begin{eqnarray*}
f_{Y|X<2}(y) &=& 
\frac{f(y \;\text{and}\; X = 0) \;+\; f(y \;\text{and}\; X = 1)}{p_0 + p_1} \\ &=&
\frac{p_0\, f_{Y|X=0}(y) \;+\; p_1\, f_{Y|X=1}(y)}{p_0 + p_1}
\end{eqnarray*}
where the second equality uses the definition of conditional density. In conclusion,
\begin{eqnarray*}
m_{Y|X<2} &=&
\int_{\infty}^{\infty} y\, \frac{p_0\, f_{Y|X=0}(y) \;+\; p_1\, f_{Y|X=1}(y)}{p_0 + p_1}\, dy \\ &=&
\frac{p_0}{p_0 + p_1} \int_{\infty}^{\infty} y\, f_{Y|X=0}(y)\, dy \;+\;
\frac{p_1}{p_0 + p_1} \int_{\infty}^{\infty} y\, f_{Y|X=1}(y)\, dy \;,
\end{eqnarray*}
that is,
$$
m_{Y|X<2} \;=\; \frac{p_0 m_0 + p_1 m_1}{p_0 + p_1} \;=\;
\frac{0.3 \cdot 6 + 0.5 \cdot 2}{0.3 + 0.5} = 3.5\;.
$$

```python
from skimage import io, color

def readGray(filename):
    img = color.rgb2gray(io.imread(filename))
    return np.around(255 * img).astype(np.uint8)

from matplotlib import pyplot as plt
%matplotlib inline

img = readGray('locomotive.jpg')
plt.imshow(img, cmap='gray')
plt.show()
```

Write a function with header

    def cumulative(img):
    
that takes an image as produced by `readGray` and computes its cumulative distribution $c(k)$. Show your code, and plot the cumulative histogram of the image in `locomotive.jpg` as read by `readGray`.

### Answer

The implementation below assumes that pixel values are integers, but does not assume a specific number of bits. It is OK if your implementation assumes `np.uint8` and hard-wires `n` to 256. However, study this code for later reference.

```python
def cumulative(img):
    assert np.issubdtype(img.dtype, np.unsignedinteger), \
        'image pixel type must be unsigned integer'
    n = np.iinfo(img.dtype).max + 1
    c = np.zeros(n)
    for value in img.flatten(): c[value] += 1
    for k in range(1, n): c[k] += c[k-1]
    for k in range(n): c[k] = c[k] / c[-1]
    return c

plt.plot(cumulative(img))
plt.show()
```

### Problem 4.5

Write a function with header

    def remap(img):
    
that takes an image as returned by `readGray` and replaces the value $k$ of every pixel with value $k' = \text{round}(255\, c(k))$, where the function $c(k)$ was defined in problem 4.3.

Show your code, display the result of running `remap` on the image in `locomotive.jpg`, and plot the cumulative histogram of the remapped image.

### Answer

The loop in the `with` construct below makes `remap` agnostic to the number of dimensions of `img`. For instance, the code would work even on a color image, or on a three-dimensional image such as a computerized tomography image. For your solution, it's OK to use nested `for` loops instead.  However, study this code for later reference. This code will only work in Python 3.6 or higher.

```python
def remap(img):
    table = cumulative(img)
    n = len(table)
    mx = n - 1
    for k in range(n):
        table[k] = round(mx * table[k])
    eq = np.zeros_like(img)
    with np.nditer([img, eq], op_flags=[['readonly'], ['writeonly']]) as idx:
        for x, y in idx: y[...] = table[x]
    return eq

crisp = remap(img)
plt.imshow(crisp, cmap='gray')
plt.show()

plt.plot(cumulative(crisp))
plt.show()
```

### Problem 4.6

Comment on the result. In particular, what would the _histogram_ (rather than the cumulative distribution) of the remapped image look like, approximately? And in what way does the remapped _image_ differ from the original, visually?

### Answer

_(This sample answer has a pedagogical purpose, and is longer and more informed than an answer that earns full credit. Parts needed for full credit are in bold.)_

The histogram $h(k)$ is the finite difference of the cumulative distribution:

$$
h(k) = \left\{\begin{array}{ll}c(0) & \mbox{for $k = 0$} \\
c(k) - c(k-1) & \mbox{otherwise.}\end{array}\right.
$$

In the case in point **the histogram would look approximately uniform**. This is why the remapping function you implemented is called _histogram equalization_. **The equalized image has** a wider dynamic range, and **greater contrast** as a result.

Histogram equalization is often used as a first stage in computer vision algorithms, because high contrast makes object recognition or other tasks easier. Of course, equalization should _not_ be used when the camera plays the role of a measurement device, that is, when the physical meaning of the pixel values is important.
